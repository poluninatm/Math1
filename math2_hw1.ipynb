### –ó–∞–¥–∞—á–∞ 2

**–ì—Ä—É–ø–ø–∞ —Å—Ç—É–¥–µ–Ω—Ç–æ–≤ –∏–∑—É—á–∞–µ—Ç 10 —Ä–∞–∑–ª–∏—á–Ω—ã—Ö –¥–∏—Å—Ü–∏–ø–ª–∏–Ω. –°–∫–æ–ª—å–∫–∏–º–∏ —Å–ø–æ—Å–æ–±–∞–º–∏ –º–æ–∂–Ω–æ —Å–æ—Å—Ç–∞–≤–∏—Ç—å —Ä–∞—Å–ø–∏—Å–∞–Ω–∏–µ –Ω–∞ –ø–æ–Ω–µ–¥–µ–ª—å–Ω–∏–∫, –µ—Å–ª–∏ –≤ —ç—Ç–æ—Ç –¥–µ–Ω—å –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å 4 —Ä–∞–∑–Ω—ã—Ö –∑–∞–Ω—è—Ç–∏—è?**

–ò—Å–ø–æ–ª—å–∑—É–µ–º —Ä–∞–∑–º–µ—â–µ–Ω–∏—è –±–µ–∑ –ø–æ–≤—Ç–æ—Ä–æ–≤, —Ç–∞–∫ –∫–∞–∫ –∑–∞–Ω—è—Ç–∏—è –†–ê–ó–ù–´–ï, –Ω–æ –ø–æ—Ä—è–¥–æ–∫ –≤–∞–∂–µ–Ω

–ß–∏—Å–ª–æ —Ä–∞–∑–º–µ—â–µ–Ω–∏–π –∏–∑ $n$ –ø–æ $k$:

$$A_n^k = \frac{n!}{(n - k)!}$$


```python
import numpy as np
A=np.math.factorial(10)/np.math.factorial(10-4)
print(f'–°—É—â–µ—Å—Ç–≤—É–µ—Ç {A} –≤–∞—Ä–∏–∞–Ω—Ç–æ–≤ —Å–æ—Å—Ç–∞–≤–ª–µ–Ω–∏—è —Ä–∞—Å–ø–∏—Å–∞–Ω–∏—è –Ω–∞ –ø–æ–Ω–µ–¥–µ–ª—å–Ω–∏–∫')
```

**–°–µ–º—å —á–µ–ª–æ–≤–µ–∫ —Ä–∞—Å—Å–∞–∂–∏–≤–∞—é—Ç—Å—è –Ω–∞—É–¥–∞—á—É –Ω–∞ —Å–∫–∞–º–µ–π–∫–µ. –ö–∞–∫–æ–≤–∞ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ –¥–≤–∞ –æ–ø—Ä–µ–¥–µ–ª—ë–Ω–Ω—ã—Ö —á–µ–ª–æ–≤–µ–∫–∞ –±—É–¥—É—Ç —Å–∏–¥–µ—Ç—å —Ä—è–¥–æ–º?**

–ß–∏—Å–ª–æ –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–æ–∫:

### –ó–∞–¥–∞—á–∞ 4

**–ò–∑ 60 –≤–æ–ø—Ä–æ—Å–æ–≤, –≤—Ö–æ–¥—è—â–∏—Ö –≤ —ç–∫–∑–∞–º–µ–Ω–∞—Ü–∏–æ–Ω–Ω—ã–µ –±–∏–ª–µ—Ç—ã, —Å—Ç—É–¥–µ–Ω—Ç –∑–Ω–∞–µ—Ç 50. –ö–∞–∫–æ–≤–∞ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ —Å—Ä–µ–¥–∏ —Ç—Ä—ë—Ö –Ω–∞—É–≥–∞–¥ –≤—ã–±—Ä–∞–Ω–Ω—ã—Ö –≤–æ–ø—Ä–æ—Å–æ–≤ —Å—Ç—É–¥–µ–Ω—Ç –∑–Ω–∞–µ—Ç: –∞) –≤—Å–µ? –±) –¥–≤–∞?**

–û–±—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–æ—á–µ—Ç–∞–Ω–∏–π –ø–æ 3 –∏–∑ 60:
    $$A = \frac{60!}{(60-3)!*3!}$$

–ò–∑ –Ω–∏—Ö –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–æ—á–µ—Ç–∞–Ω–∏–π —Å —Ç–µ–º–∏ –≤–æ–ø—Ä–æ—Å–∞–º–∏, –∫–æ—Ç–æ—Ä—ã–µ –æ–Ω –∑–Ω–∞–µ—Ç: 
    $$An = \frac{50!}{(50-3)!*3!}$$

```python
A=np.math.factorial(60)/(np.math.factorial(57) * np.math.factorial(3))
An=np.math.factorial(50)/(np.math.factorial(47) * np.math.factorial(3))
print(f'–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ —Å—Ç—É–¥–µ–Ω—Ç –≤—ã—Ç—è–Ω–µ—Ç —Ç—Ä–∏ –∏–∑–≤–µ—Å—Ç–Ω—ã—Ö –µ–º—É –≤–æ–ø—Ä–æ—Å–∞: {An/A}')
```

–†–∞–∑–ª–∏—á–Ω—ã–µ —Å–æ—á–µ—Ç–∞–Ω–∏—è –∏–∑–≤–µ—Å—Ç–Ω—ã—Ö –±–∏–ª–µ—Ç–æ–≤ –∏ –Ω–µ–∏–∑–≤–µ—Å—Ç–Ω–æ–≥–æ –º–æ–∂–Ω–æ –ø–æ–ª—É—á–∏—Ç—å, –ø–µ—Ä–µ–º–Ω–æ–∂–∏–≤ —Å–æ—á–µ—Ç–∞–Ω–∏—è –¥–ª—è –∏–∑–≤–µ—Å—Ç–Ω—ã—Ö –∏ –Ω–µ–∏–∑–≤–µ—Å—Ç–Ω–æ–≥–æ –±–∏–ª–µ—Ç–æ–≤: 
    $$An = \frac{50!}{(50-2)!*2!}$$
    $$Au = \frac{10!}{(10-1)!*1!}$$
–ò –ø–æ–ª—É—á–∏—Ç—å –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ –æ–¥–∏–Ω –∏–∑ –≤–æ–ø—Ä–æ—Å–æ–≤ —Å—Ç—É–¥–µ–Ω—Ç—É –±—É–¥–µ—Ç –Ω–µ–∏–∑–≤–µ—Å—Ç–µ–Ω:
     $$P_2 = \frac{(An*Au)}{A}$$

```python
A=np.math.factorial(60)/(np.math.factorial(57) * np.math.factorial(3))
An=np.math.factorial(50)/(np.math.factorial(50-2) * np.math.factorial(2))
Au=np.math.factorial(10)/(np.math.factorial(10-1) * np.math.factorial(1))
print(f'–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ —Å—Ç—É–¥–µ–Ω—Ç –≤—ã—Ç—è–Ω–µ—Ç —Ç–æ–ª—å–∫–æ –¥–≤–∞ –∏–∑–≤–µ—Å—Ç–Ω—ã—Ö –µ–º—É –≤–æ–ø—Ä–æ—Å–∞: {(An*Au)/A}')
```

**–ë—Ä–æ—Å–∞–µ—Ç—Å—è –∏–≥—Ä–∞–ª—å–Ω–∞—è –∫–æ—Å—Ç—å. –ü—É—Å—Ç—å —Å–æ–±—ã—Ç–∏–µ A - –ø–æ—è–≤–ª–µ–Ω–∏–µ —á—ë—Ç–Ω–æ–≥–æ —á–∏—Å–ª–∞, —Å–æ–±—ã—Ç–∏–µ B - –ø–æ—è–≤–ª–µ–Ω–∏–µ —á–∏—Å–ª–∞ –±–æ–ª—å—à–µ —Ç—Ä—ë—Ö. –Ø–≤–ª—è—é—Ç—Å—è –ª–∏ —ç—Ç–∏ —Å–æ–±—ã—Ç–∏—è –Ω–µ–∑–∞–≤–∏—Å–∏–º—ã–º–∏?**

–°–æ–±—ã—Ç–∏—è —è–≤–ª—è—é—Ç—Å—è –∑–∞–≤–∏—Å–∏–º—ã–º–∏ - —Ç–∫ –Ω–∞—Å—Ç—É–ø–ª–µ–Ω–∏–µ –æ–¥–Ω–æ–≥–æ (–ø–æ—è–≤–ª–µ–Ω–∏–µ —á–µ—Ç–Ω–æ–≥–æ —á–∏—Å–ª–∞) –∏–∑–º–µ–Ω—è–µ—Ç –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –≤—Ç–æ—Ä–æ–≥–æ (—Ç–æ, —á—Ç–æ —ç—Ç–æ —á–∏—Å–ª–æ –±–æ–ª—å—à–µ —Ç—Ä–µ—Ö)

### –ó–∞–¥–∞—á–∞ 6

**–î–æ–ø—É—Å—Ç–∏–º, –∏–º–µ–µ—Ç—Å—è –Ω–µ–∫–æ—Ç–æ—Ä–∞—è –æ—á–µ–Ω—å —Ä–µ–¥–∫–∞—è –±–æ–ª–µ–∑–Ω—å (–ø–æ—Ä–∞–∂–∞–µ—Ç 0.1% –Ω–∞—Å–µ–ª–µ–Ω–∏—è). –í—ã –ø—Ä–∏—Ö–æ–¥–∏—Ç–µ –∫ –≤—Ä–∞—á—É, –≤–∞–º –¥–µ–ª–∞—é—Ç —Ç–µ—Å—Ç –Ω–∞ —ç—Ç—É –±–æ–ª–µ–∑–Ω—å, –∏ —Ç–µ—Å—Ç –æ–∫–∞–∑—ã–≤–∞–µ—Ç—Å—è –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–º. –í—Ä–∞—á –≥–æ–≤–æ—Ä–∏—Ç –≤–∞–º, —á—Ç–æ —ç—Ç–æ—Ç —Ç–µ—Å—Ç –≤–µ—Ä–Ω–æ –≤—ã—è–≤–ª—è–µ—Ç 99 % –±–æ–ª—å–Ω—ã—Ö —ç—Ç–æ–π –±–æ–ª–µ–∑–Ω—å—é –∏ –≤—Å–µ–≥–æ –ª–∏—à—å –≤ 1% —Å–ª—É—á–∞–µ–≤ –¥–∞—ë—Ç –ª–æ–∂–Ω—ã–π –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–π –æ—Ç–≤–µ—Ç.**

–í–æ–ø—Ä–æ—Å: –≤–∞–∫–æ–≤–∞ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å, —á—Ç–æ –≤—ã –¥–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ –±–æ–ª—å–Ω—ã –µ–π?

–ü–æ–¥—Å–∫–∞–∑–∫–∞: –∏—Å–ø–æ–ª—å–∑—É–π—Ç–µ —Ñ–æ—Ä–º—É–ª—É –ë–∞–π–µ—Å–∞ —Å —Ä–∞—Å–∫—Ä—ã—Ç–∏–µ–º –∑–Ω–∞–º–µ–Ω–∞—Ç–µ–ª—è —Å –ø–æ–º–æ—â—å—é —Ñ–æ—Ä–º—É–ª—ã –ø–æ–ª–Ω–æ–π –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏.

–ï—Å–ª–∏ –í - –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –∑–∞–±–æ–ª–µ—Ç—å —ç—Ç–æ–π –±–æ–ª–µ–∑–Ω—å—é, —Ç–æ A –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–≥–æ —Ç–µ—Å—Ç–∞ –Ω–∞ —ç—Ç—É –±–æ–ª–µ–∑–Ω—å

$$P(B\:|\:A) = \frac{P(A\:|\:B) \cdot P(B)}{P(A)}$$

–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å $${P(A\:|\:B)= 0.99}$$

$${P(B)= 0.001}$$

–ü–æ–ª–Ω–∞—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –¥–ª—è A: 
$${P(A)= 0.99*0.001+0.01*(1-0.001)}$$

```python
# ùëÉ(ùê¥)
0.99*0.001+0.01*(1-0.001)
```
