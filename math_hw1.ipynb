### 5. Задание (в программе)
1) Нарисуйте трехмерный график двух параллельных плоскостей.

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)

x = np.arange(-5, 5, 0.5) 
y = np.arange(-5, 5, 0.5)

x , y, =np.meshgrid(x, y)
z = 2*x+3*y+10
ax.plot_wireframe(x, y, z)
z2 = 2*x+3*y+30
ax.plot_wireframe(x, y, z2, color="red")
show()
```

2) Нарисуйте трехмерный график двух любых поверхностей второго порядка.

```python
x = np.arange(-5, 5, 0.5) 
y = np.arange(-5, 5, 0.5)

fig = figure()
ax = Axes3D(fig)

x , y, =np.meshgrid(x, y)
z = x**2+y**2
ax.plot_surface(x, y, z)

z2 = np.sin(x**2+y**2)
ax.plot_surface(x, y, z2, color="red")
show()
```
